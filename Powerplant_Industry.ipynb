```python
#preprocessing (Feature scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scala = df_new[['capacity_mw','generation_gwh_2016','generation_gwh_2017','estimated_generation_gwh_2016']]
#df_scala = pd.DataFrame(scaler.fit_transform(df_scala), columns = df_scala.columns)
df_scala = pd.DataFrame(df_scala, columns = df_scala.columns)
```

```python
#Training the dataset
from sklearn.model_selection import train_test_split

X = df_scala[['capacity_mw','generation_gwh_2016','generation_gwh_2017']]
y = df_scala['estimated_generation_gwh_2016']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
#80% training and 20% testing
```

```python
#Testing and finding R2 score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
rfr = RandomForestRegressor().fit(X_train, y_train)

y_pred = rfr.predict(X_test)

rfr_score = r2_score(y_test, y_pred)  #comparing the predicted value and value in dataset
rfr_rmse = mean_squared_error(y_test, y_pred, squared = False)
print("R2 Score : ", rfr_score)
print("RMSE : ", rfr_rmse)
```

```python
#Giving our input in the form of csv and getting the predicted results

import math
my=pd.read_csv("D:\\Documents\\user_input_powerplant.csv")
my_pred=rfr.predict(my)
# print(sum(my_pred))
pred_my=sum(my_pred)
gwh_to_kwh=pred_my*math.pow(10,6)
kwh_to_CO2_kg=0.23314*gwh_to_kwh
CO2_in_lbs=kwh_to_CO2_kg*2.20462
trees=math.ceil(CO2_in_lbs/50)
print(f"Net amount of CO2 released in your area is (in lbs CO2) {CO2_in_lbs}")
print(f"The no.of trees required to offset the CO2 emission is : {trees}")
# new_data=pd.DataFrame(data=my_pred) 
# print(new_data)
```
